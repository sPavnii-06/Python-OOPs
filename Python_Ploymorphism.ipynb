# Polymorphism in OOP Python
Polymorphism is one of the key principles of Object-Oriented Programming (OOP) that allows a single interface to represent different underlying forms (data types).
It enables objects of different classes to be treated as objects of a common super class.

### 1. Concept of Polymorphism
- Polymorphism means "many forms." 
- In Python OOP, it allows different classes to have methods with the same name but different implementations.
- The main idea is that a single function or method can work with objects of multiple types, making code more flexible and reusable.

### 2. Types of Polymorphism in Python

Polymorphism in Python can be categorized into the following types:

- A. Compile-Time Polymorphism (Method Overloading)
Python does not support traditional method overloading like Java or C++.

However, we can achieve a similar effect using default arguments or variable-length arguments (*args, **kwargs).

#### Example of Method Overloading in Python

```python
class Math:
    def add(self, a, b, c=0):  # Default argument
        return a + b + c

M= Math()
print(M.add(2, 3))       # Output: 5
print(M.add(2, 3, 4))    # Output: 9
```

### B. Runtime Polymorphism (Method Overriding)
Happens when a subclass provides a specific implementation of a method that is already defined in its superclass.

#### Example of Method Overriding

```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):
        return "Bark"

class Cat(Animal):
    def speak(self):
        return "Meow"

animals = [Dog(), Cat()]

for animal in animals:
    print(animal.speak())
```

### C. Polymorphism with Functions and Objects

A function can work with multiple types of objects if they share a common interface.

#### Example of Polymorphism with Functions

```python
class Bird:
    def fly(self):
        return "Bird can fly"

class Airplane:
    def fly(self):
        return "Airplane can fly"

def flying(obj):
    print(obj.fly())

# Using polymorphism
bird = Bird()
plane = Airplane()

flying(bird)   # Output: Bird can fly
flying(plane)  # Output: Airplane can fly
```

### D. Operator Overloading (Ad-hoc Polymorphism)
- In Python, we can redefine the behavior of operators for user-defined objects by overloading special methods (e.g., __add__, __mul__).
- Operator overloading is a feature in Python that allows us to redefine the behavior of built-in operators (+, -, *, /, ==, etc.) for user-defined objects. This enables objects of a class to behave in a polymorphic manner when used with operators.

- Polymorphism Aspect: The same operator (+, -, *, etc.) can perform different operations depending on the object types.

### Why Use Operator Overloading?

- Makes custom objects behave like built-in types.

- Improves code readability and usability.

- Allows intuitive mathematical and logical operations on user-defined objects.



### How to Overload Operators?

- Python provides special methods (magic methods or dunder methods) to overload operators.
   
- These methods start and end with double underscores (__).

  

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):  # Overloading `+` operator
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # Calls __add__
print(v3)  # Output: (6, 8)
```

### Summary

- Polymorphism allows a common interface for multiple object types.

- Python supports method overriding, operator overloading, and function polymorphism.

- It enhances code readability, maintainability, and flexibility.

- Used in object-oriented design, inheritance, and dynamic method resolution.
