### Static and Final Keyword, Access Modifiers and specifiers, scope of a class in python

### 1. static and final Keywords in Python

- Python does not have explicit static or final keywords like Java, but similar behavior can be achieved:

#### Static Variables and Methods
Static variables and static methods belong to the class itself rather than an instance of the class. This means:

Static Variables are shared across all instances of a class.

Static Methods can be called on the class itself and do not require an instance.

   - In Python, static variables and methods are defined using the @staticmethod decorator.
   - Static methods do not access instance attributes (self) or class attributes (cls).

```python
class Example:
    static_var = "I am static"

    @staticmethod
    def static_method():
        print("This is a static method")

# Accessing without creating an instance

print(Example.static_var)  # Output: I am static

Example.static_method()    # Output: This is a static method
```

### Final Variables and Methods

- Python does not have a final keyword, but similar behavior can be achieve:

#### Final Variables: Use constants (by convention, uppercase names indicate constants).

#### Final Methods and Classes: Use the @final decorator from typing.

```python
from typing import final

class Base:
    @final
    def cannot_override(self):
        print("This method cannot be overridden")

@final
class Derived(Base):
    pass

# class Another(Derived):  # This will raise an error
#     pass
```

### 2. Access Modifiers in Python
Unlike Java, Python does not enforce access modifiers strictly. Instead, it follows naming conventions:



```python
class AccessModifiers:
    public_var = "I am Public"
    _protected_var = "I am Protected"
    __private_var = "I am Private"

    def __init__(self):
        print(self.public_var)
        print(self._protected_var)
        print(self.__private_var)

obj = AccessModifiers()
print(obj.public_var)        # Works
print(obj._protected_var)    # Works but should be avoided
# print(obj.__private_var)   # AttributeError: object has no attribute '__private_var'

print(obj._AccessModifiers__private_var)  # Works due to name mangling
```

### 3. Scope of a Class in Python
- The scope of a class refers to where it can be accessed in the program.

#### Global Scope: 
     Classes defined at the top level of a module are available throughout the module.

#### Local Scope: 
     Classes defined inside a function are accessible only within that function.

#### Nested Classes: 
     Inner classes can be accessed only through their enclosing class.

## Global Scope Example

```python
class GlobalClass:
    pass  # Can be used anywhere in the module
```

## Local Scope Example

```python
def my_function():
    class LocalClass:
        pass

    obj = LocalClass()  # Works inside function

#obj = LocalClass()  # Error: LocalClass is not defined outside
```

## Nested Class Example

```python
class Outer:
    class Inner:
        def display(self):
            print("Inside Inner Class")

    def outer_method(self):
        obj = self.Inner()
        obj.display()

o = Outer()
o.outer_method()
```
