# Inheritence

### 1. Single Level Inheritence

```python
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def show_person_details(self):
        print(f"{self.name},{self.age},{self.gender}")

class Student(Person):
    def __init__(self,name,age,gender,sysid,course,dept):
        super().__init__(name,age,gender)
        self.sysid=sysid
        self.course=course
        self.dept=dept

    def show_student_details(self):
        self.show_person_details()
        print(f"{self.sysid},{self.course},{self.dept}")
        
```

### 2.Multiple Inheritence 

```python
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def show_person_details(self):
        print(f"{self.name},{self.age},{self.gender}")

class Student:
    def __init__(self,sysid,course,dept):
        self.sysid=sysid
        self.course=course
        self.dept=dept

    def show_student_details(self):
        print(f"{self.sysid},{self.course},{self.dept}")

class Employee(Person,Student):
    def __init__(self,name,age,gender,sysid,course,dept,empid,salary,exp):
        #super().__init__(name,age,gender)
        #Student.__init__(self,sysid,course,dept)
        Person.__init__(self,name,age,gender)
        Student.__init__(self,sysid,course,dept)
        self.empid=empid
        self.salary=salary
        self.exp=exp

    def show_emp(self):
        self.show_person_details()
        self.show_student_details()
        print(f"{self.empid},{self.salary},{self.exp}")
        
```

```python
class Person:
    def __init__(self,nm,age,gender):
        self.nm=nm
        self.age=age
        self.gender=gender
    def show_person_details(self):
        print(f"{self.nm},{self.age},{self.gender}")

class Student:
    def __init__(self,sysid,course,dept):
        self.sysid=sysid
        self.course=course
        self.dept=dept

    def show_student_details(self):
        print(f"{self.sysid},{self.course},{self.dept}")

class Employee(Person,Student):
    def __init__(self,nm,age,gender,sysid,course,dept,empid,salary,exp):
        Person.__init__(self,nm,age,gender)
        Student.__init__(self,sysid,course,dept)
        self.empid=empid
        self.salary=salary
        self.exp=exp

    def show_emp(self):
        Person.show_person_details(self)
        Student.show_student_details(self)
        print(f"{self.empid},{self.salary},{self.exp}")
```

### 3. Multi-level Inheritence

```python
# Base Class
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def show_personal_info(self):
        print(f"{self.name},{self.age},{self.gender}")
```

```python
# Derived Class
class Registration(Person):
    def __init__(self,name,age,gender,rid,course,fee):
        super().__init__(name,age,gender)
        self.rid=rid
        self.course=course
        self.fee=fee

    def show_registration(self):
        self.show_personal_info()
        print(f"{self.rid},{self.course},{self.fee}")  
```

```python
# Derived Class 
class Student(Registration):
    def __init__(self,name,age,gender,rid,course,fee,sysid,marks):
        super().__init__(name,age,gender,rid,course,fee)
        self.sysid=sysid
        self.marks=marks
    def show_student_details(self):
        self.show_registration()
        print(f"{self.sysid},{self.marks}")               
```

```python
stu=Student("Rama",20,"M",123,"BTECH",500000,"E1",44)
stu.show_student_details()
```
