# Classes and Objects without constructor

### Class

```python
class Student:
    def get_details(self,system_id,name,course,marks):
        self.system_id=system_id
        self.name=name
        self.course=course
        self.marks=marks
    def show_details(self):
        print(f"{self.system_id},{self.name},{self.course},{self.marks}")
```

### Objects

```python
s1=Student()
s1.get_details(1,"Abhi","BTECH",40)
s1.show_details()
```

# Inheritence

### 1. Single Inheritence

```python
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def show_person_details(self):
        print(f"{self.name},{self.age},{self.gender}")

class Student(Person):
    def __init__(self,name,age,gender,sysid,course,marks):
        super().__init__(name,age,gender)
        self.sysid=sysid
        self.course=course
        self.marks=marks

    def show_student_details(self):
        print(f"{self.sysid},{self.course},{self.marks}")
        
```

```python
st=Student("Abhi",20,"M","S1","BTech",40)
st.show_person_details()
st.show_student_details()
```

```python
st=Student("Abhi",20,"M","S1","BTech",40)
st.show_details()
```

### 2. Multiple Inheritence

```python
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def show_person_details(self):
        print(f"{self.name},{self.age},{self.gender}")

class Student:
    def __init__(self,sysid,course,marks):
        self.sysid=sysid
        self.course=course
        self.marks=marks

    def show_stu_details(self):
        print(f"{self.sysid},{self.course},{self.marks}")
        
```

```python
class Employee(Person,Student):
    def __init__(self,name,age,gender,sysid,course,marks,empid,salary,exp):
        Person.__init__(self,name,age,gender)
        Student.__init__(self,sysid,course,marks)
        self.empid=empid
        self.salary=salary
        self.exp=exp
    def show_emp_details(self):
        Person.show_person_details(self)
        Student.show_stu_details(self)
        print(f"{self.empid},{self.salary},{self.exp}")
```

```python
emp=Employee("Swati",20,"F","S121","BTech",40,"E12",50000,3)
emp.show_emp_details()
```

### 3. Multi-Level Inheritence

```python
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def show_person_details(self):
        print(f"{self.name},{self.age},{self.gender}")
```

```python
class Registration(Person):
    def __init__(self,name,age,gender,rid,rfee):
        super().__init__(name,age,gender)
        self.rid=rid
        self.rfee=rfee
    def show_registration(self):
        print(f"{self.rid},{self.rfee}")
        print("Registration Done")

class Student(Registration):
    def __init__(self,name,age,gender,rid,rfee,sysid,course,marks):
        super().__init__(name,age,gender,rid,rfee)
        self.sysid=sysid
        self.course=course
        self.marks=marks

    def show_stu_details(self):
        print(f"{self.sysid},{self.course},{self.marks}")
        print("The Details of students are shown above")
        
```

```python
st1=Student("Ravi",20,"M","R1",5000000,"ST1","Btech",40)
st1.show_person_details()
st1.show_registration()
st1.show_stu_details()
```
